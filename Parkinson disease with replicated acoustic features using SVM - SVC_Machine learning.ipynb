```python
y = dt['Status']
y.head()
```

```python
X = dt.drop(columns = ['Status'])
X.head()
```

```python
X.shape

```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 4)
pca.fit(X)
X = pca.transform(X)
```

```python
X.shape
```

```python
fig = plt.plot(X)
plt.legend(fig)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X
```

```python
plt.plot(X)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test = train_test_split(X, y , random_state = 20, test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import plot_confusion_matrix
svc = SVC(random_state = 20)
svc.fit(X_train,Y_train)
```

```python
plot_confusion_matrix(svc,X_test,Y_test,values_format = 'd')
```

```python
from sklearn.model_selection import GridSearchCV

para_grid = [ { 'C': [0, 0.25,0.5,1,10,100],
            'gamma' : ['scale',1,0.1,0.001,0.0001,0.00001],
            'kernel': ['rbf'] }, ]
optimize_param = GridSearchCV(SVC(),para_grid,cv = 10, scoring = 'accuracy', verbose = 2)
```

```python
optimize_param.fit(X_train,Y_train)
```

```python
print(optimize_param.best_params_)
```

```python
svc1 = SVC(random_state = 20, kernel = 'rbf', C = 100, gamma = 0.001)
svc1.fit(X_train,Y_train)
```

```python
plot_confusion_matrix(svc1,X_test,Y_test,values_format = 'd')
```

```python
y_pred = svc1.predict(X_test)
```

```python
plot_confusion_matrix(svc1,X_test,y_pred,values_format = 'd')
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
print('The mse is :',mean_squared_error(Y_test,y_pred))
print('The cr is{}'.format(classification_report(Y_test,y_pred)))
print('Accuracy:' , accuracy_score(Y_test,y_pred) )
confusion_matrix(Y_test,y_pred, sample_weight = None)
```
